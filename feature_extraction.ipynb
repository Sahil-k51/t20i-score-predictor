```python
cities = np.where(df['city'].isnull(),df['venue'].str.split().apply(lambda x:x[0]),df['city'])
```

```python
df['city'] = cities
```

```python
df.drop(columns=['venue'], inplace = True)
```

```python
eligible_cities = df['city'].value_counts()[df['city'].value_counts() > 600].index.tolist()
```

```python
df[df['city'].isin(eligible_cities)]
```

```python
# Convert 'runs' column to a numeric type (e.g., int) before applying cumsum
df['runs'] = pd.to_numeric(df['runs'], errors='coerce')  # 'coerce' handles non-numeric values as NaN

# Now, calculate the cumulative sum
df['current_score'] = df.groupby('match_id')['runs'].cumsum()
```

```python
df
```

```python
df['over'] = df['ball'].apply(lambda x:str(x).split(".")[0])
df['ball_no'] = df['ball'].apply(lambda x:str(x).split(".")[1])
df
```

```python
df['balls_bowled'] = (df['over'].astype('int')*6) + df['ball_no'].astype('int')
df
```

```python
df['balls_left'] = 120 - df['balls_bowled']
df['balls_left'] = df['balls_left'].apply(lambda x:0 if x<0 else x)
df
```

```python
# Convert 'player_dismissed' to numeric, handling errors
df['player_dismissed'] = pd.to_numeric(df['player_dismissed'], errors='coerce')  

# Calculate cumulative sum after grouping
df['player_dismissed'] = df.groupby('match_id')['player_dismissed'].cumsum()

# Fill any missing values (NaN) with 0
df['player_dismissed'] = df['player_dismissed'].fillna(0).astype('int')  

df['wickets_left'] = 10 - df['player_dismissed']
```

```python
df
```

```python
df['crr'] = (df['current_score']*6)/df['balls_bowled']
```

```python
# Convert 'runs' column to numeric, handling errors
df['runs'] = pd.to_numeric(df['runs'], errors='coerce') 

# Fill any missing values with 0
df['runs'] = df['runs'].fillna(0).astype('int') 

groups = df.groupby('match_id')

match_ids = df['match_id'].unique()
last_five = []
for id in match_ids:
    # Select only the 'runs' column before calculating the rolling sum
    last_five.extend(groups.get_group(id)['runs'].rolling(window=30).sum().values.tolist())
```

```python
df
```

```python
df['last_five'] = last_five
df
```

```python
final_df = df.groupby('match_id').sum()['runs'].reset_index().merge(df,on='match_id')
```

```python
final_df=final_df[['batting_team','bowling_team','city','current_score','balls_left','wickets_left','crr','last_five','runs_x']]
```

```python
final_df.dropna(inplace=True)
```

```python
final_df.isnull().sum()
```

```python
final_df = final_df.sample(final_df.shape[0])
```

```python
final_df.sample(2)
```

```python
X = final_df.drop(columns=['runs_x'])
y = final_df['runs_x']
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
X_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
trf = ColumnTransformer([
    ('trf',OneHotEncoder(drop='first'),['batting_team','bowling_team','city'])
]
,remainder='passthrough')
```

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',StandardScaler(with_mean=False)),
    ('step3',XGBRegressor(n_estimators=1000,learning_rate=0.2,max_depth=12,random_state=1))
])
```

```python
pipe.fit(X_train, y_train)
y_pred = pipe.predict(X_test)
print(r2_score(y_test, y_pred))
print(mean_absolute_error(y_test, y_pred))
```

```python
pickle.dump(pipe, open('pipe.pkl', 'wb'))
```
